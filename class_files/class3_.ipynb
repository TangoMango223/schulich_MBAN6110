```python
# Map function to map the average price to each column
df["avg_price_per_sqft"] = df["zipcode"].map(price_per_sqft)

# It must be in a dictionary format

# The alternative is join or merge in a new frame
```

```python
# store
price_per_sqft_frame = df.groupby("zipcode")["price_per_sqft"].mean().to_frame()

price_per_sqft_frame
```

```python
new_df = pd.merge(df, price_per_sqft_frame, left_on = "zipcode", right_on = "zipcode")

new_df
```

```python
# Is price above/below the average for zipcode
df["above/below"] = df["price_per_sqft"] - df["avg_price_per_sqft"]
```

```python
# Average sqft 
df["avg_sqft_for_area"] = df["zipcode"].map(df.groupby("zipcode")["sqft_living"].mean().to_dict())
```

```python
# Above/Below
df["above/below_sqft"] = df["sqft_living"] - df["avg_sqft_for_area"]
```

```python
# Refine your definition of a good sale

# Super common for targetting users, customers, etc. How many people have done, something, etc. etc.
# You might have slightly different values of each time

# Custom function

def sale_category(row):
    # Condition #1 - house sold for less per sqft than average price for zipcode, and house was bigger than avg. size.
    # Then, we have a deal
    if row["above/below"] < 0 and row['above/below_sqft'] > 0:
        return "Value Sale"
    # Condition #2 - house sold for less per sqft than avg price for zipcode, or house was bigger than avg for the area
    # Not a deal, but good outcome
    elif row["above/below"] < 0 or row['above/below_sqft'] > 0:
        return "average sale"
    # If none of these, then house was overpriced
    else:
        return "Overpriced"
            

```

```python
# Apply function - apply a custom treatment
df["sale_value"] = df.apply(sale_category, axis = 1) # apply treatment by row

df.head(5)
```

```python
# Missing Data:

# Why is the data missing? Three options why it is missing

# MCAR, MAR, MNAR

# Missing Completely at random

# Missing At Random

# Data Missing Not at Random

# Might actually add a charactierstics

# Data Missing at Random - no specific reason - in theory everyone has an equal likelihood to leave a review, but generally it's in extremes
# There is a PATTERN with the data missing

# Missing completely at random - MCAR
# Ask people to give you their DOB, and it's a completely random sample, 1000 users of different people from the world
# Some share, some do not - people who do not celebrate their birthday or people don't feel comfortable

# If data is compeltely missing at random, then no matter how you're doing it, you are not introducing any bias

# String or invisible character --> check dtype, it should say object if there's any, when it should be fully numeric

# Missing data not at random
# Missing data does not need to be filled, incomplete cohort (everyone who signed up on April 15th)
# Everyone there only has 15 days of data, not the 45 days since they haven't been here for long enough.
# Sometimes people may have a choice, such as optional
# Problem with a data measurement mechanism. let's say Facebook asked you for your salary - problem with how the question is being asked.
```
