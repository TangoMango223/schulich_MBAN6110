```python
# T-Test
store_a_sales = np.random.normal(30000, 5000, 100) # average is 30K
store_b_sales = np.random.normal(35000, 5000, 100) # average is 35K

# Check statistical significance between 2 means of the same thing, i.e. Male and Female GMAT test scores
```

```python
# Import
from scipy.stats import ttest_ind
# Set equl_var = False - the test assumes there is equal variance, you might set to true if you don't know
```

```python
# Calculate mean
mean_a = sum(store_a_sales) / len(store_a_sales)
mean_b = sum(store_b_sales) / len(store_b_sales)

print(mean_a)
print(mean_b)
```

```python
# T-Test

# 2-tail

# if you test greater than for Store B > Store A - 
t_stat, p_value = ttest_ind(store_a_sales, store_b_sales, equal_var = False, alternative= 'less')
print(p_value)
```

```python
# One-sided - p = 0.999 - GREATER

# Very high p-value, so not statistical significant differences in means.
# Store B is not actually larger than store A
# Bold assumption that data is normally distributed

# It changes the underlying formula

# Two-tailed, p < 0.05
# it's not greater, so the means are actually statistically different

# One-tail, less, p < 0.05
# Store is actually
```

```python
# Mann-Whitney Test

# Non-Parametric - not normally distributed
from scipy.stats import mannwhitneyu
```

```python
# code - default is two sided
u_stat, p_value = mannwhitneyu(store_a_sales, store_b_sales , alternative = "less")
```

```python
print(p_value)

# statistically significant

# better for non-normal data
# but if it's normal, it's less precise than t-test
# It is much more normal
```

```python
# if you plan to do testing, it takes a very long time to run an experiment for it to be converging normal
# non-parametric testing is more popular

# You can statistically significant results - but T-Test tells you otherwise
```
