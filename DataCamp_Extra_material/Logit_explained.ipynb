# Logistic Regression Explained for Dummies

Note:

* Linear Combination to calculate some score, with input vatiables to calculate "z"
* z is then passed into a Sigmoid function to produce a probability
* Some threshold is prov ided on whether

```python
import numpy as np
import pandas as pd
from statsmodels.formula.api import logit
import seaborn as sns
import matplotlib.pyplot as plt

# Simulated dataset
data = pd.DataFrame({
    'x': np.linspace(0, 10, 100),
    'y': np.random.binomial(1, p=0.5, size=100)
})

# Fit logistic regression model
model = logit('y ~ x', data=data).fit()
print(model.summary())

# Plotting the logistic regression model
sns.regplot(x='x', y='y', data=data, ci=None, logistic=True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import logit

# Step 1: Generate Synthetic Data
np.random.seed(0)
n_samples = 100
x = np.linspace(0, 10, n_samples)
# Create a logistic function
p = 1 / (1 + np.exp(-(x - 5)))
# Generate binary outcomes
y = np.random.binomial(1, p)

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Step 2: Fit the Logistic Regression Model
model = logit('y ~ x', data=data).fit()
print(model.summary())

# Step 3: Visualize the Data and Fitted Model
# Plot the data
sns.scatterplot(x='x', y='y', data=data)

# Plot the fitted logistic regression curve
sns.regplot(x='x', y='y', data=data, ci=None, logistic=True, line_kws={'color': 'red'})

plt.show()
```
